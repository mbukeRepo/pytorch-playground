```python
class Dataset(Dataset):
    def __init__(self, x, y):
        self.x  = x
        self.y = y
    def __getitem__(self, index):
        return self.x[index], self.y[index]
    def __len__(self):
        return len(self.x)
```

```python
dataset = Dataset(x, y)
```

```python
len(dataset)
```

```python
train_loader = torch.utils.data.DataLoader(dataset = dataset, batch_size= 32, shuffle=True)
```

```python
train_loader
```

```python
class Model(nn.Module):
    def __init__(self, input_features, output_features):
        super(Model, self).__init__()
        self.fc1 = nn.Linear(input_features, 5)
        self.fc2 = nn.Linear(5, 4)
        self.fc3 = nn.Linear(4, 3)
        self.fc4 = nn.Linear(3, output_features)
        self.sigmoid = nn.Sigmoid()
        self.tanh = nn.Tanh()

    def forward(self, x):
        out = self.fc1(x)
        out = self.tanh(out)
        out = self.fc2(out)
        out = self.tanh(out)
        out = self.fc3(out)
        out = self.tanh(out)
        out = self.fc4(out)
        out = self.sigmoid(out)

        return out
    
    
```

```python
net = Model(7, 1)
```

```python
loss_function = torch.nn.BCELoss(size_average=True)
```

```python
optimizer = torch.optim.SGD(net.parameters(), lr = 0.1, momentum = 0.9)
```

```python
# train the network
```

```python
# parameters
```

```python
epochs = 200
```

```python
for epoch in range(epochs):
    for (input, labels) in train_loader:
        input = input.float()
        labels = labels.float()
        # forward the net
        output = net(input)
        loss = loss_function(output, labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
    # accuracy calculation  => Thresholding 
    output = (output > 0.5).float()
    accuracy = (output == labels).float().mean()
    print("Epoch {}/{}: Loss: {} Accuracy: {:.3f}".format(epoch + 1, 200, loss, accuracy))
```
